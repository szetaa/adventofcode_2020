```python
with open('2020-12-02_input.txt') as f:
    records = f.read().splitlines()
```

```python
records[0:3]
```

## First part

```python
import re
normalized = []
valid = []
for r in records:
    po, pw = r.split(':')
    lo, up = re.findall(r'\d+',po)
    lo=int(lo)
    up=int(up)
    le = re.findall(r'[a-z]',po)[0]
    co = pw.count(le)
    if ((co >= lo) and (co <= up)):
        valid.append([lo,up,le,pw,co])
    normalized.append([lo,up,le,pw,co])
    
```

```python
len(valid)
```

## Second part

```python
import re
normalized = []
valid = []
for r in records:
    po, pw = r.split(':')
    pw=pw[1:]
    lo, up = re.findall(r'\d+',po)
    lo=int(lo)
    up=int(up)
    le = re.findall(r'[a-z]',po)[0]
    lox = pw[lo-1]==le
    upx = pw[up-1]==le
    if (lox+upx==1):
        valid.append([lo,up,le,pw])
    normalized.append([lo,up,le,pw])
    
```

```python
len(valid)
```
